```python
import os
import sys
import cv2
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

sys.path.insert(0, os.path.abspath(os.path.join('..')))
from util import plot_image_with_boxes
```

```python
root_dir_obj_det = "/Stor1/wout/TreeAI4Species/ObjDet/"
dir_objdet = os.path.join(root_dir_obj_det, "12_RGB_ObjDet_640_fL")

root_dir_sem_seg = "/Stor1/wout/TreeAI4Species/SemSeg/"
dir_semseg = os.path.join(root_dir_sem_seg, "12_RGB_SemSegm_640_fL")
```

```python
# objdet_id = "000000002854"
objdet_id = "000000002980"
label_file = os.path.join(dir_objdet, "train", "labels", f"{objdet_id}.txt")
img_file = os.path.join(dir_objdet, "train", "images", f"{objdet_id}.png")

plot_image_with_boxes(img_file, label_file)

# plot semseg image with mask
# semseg_id = "000000002234"
semseg_id = "000000002370"
label_file = os.path.join(dir_semseg, "train", "labels", f"{semseg_id}.png")
img_file = os.path.join(dir_semseg, "train", "images", f"{semseg_id}.png")

image = cv2.imread(img_file)
with Image.open(label_file) as m:
    mask = np.array(m).astype(np.int64)

f, axs = plt.subplots(1,2, sharey=True, figsize=(12,6))
axs[0].imshow(image)
axs[1].imshow(mask)

print(f"Unique labels in mask: {np.unique(mask)}")
```
